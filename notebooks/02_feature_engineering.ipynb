```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Urvi Patel\customer-intelligence-system\data\processed\cleaned_telco.csv")
```

```python
df['Churn'].unique()
```

```python
def tenure_group(x):
    if x <= 12:
        return "0-1_year"
    elif x <= 24:
        return "1-2_year"
    elif x <= 48:
        return "2-4_year"
    else:
        return "4+_year"

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

```python
df['avg_charges'] = df['TotalCharges'] / (df['tenure'] + 1)
```

```python
services = [
    'OnlineSecurity','OnlineBackup','DeviceProtection',
    'TechSupport','StreamingTV','StreamingMovies'
]

df['num_services'] = df[services].apply(lambda x: sum(x == 'Yes'), axis=1)
```

```python
df['high_value'] = (df['MonthlyCharges'] > df['MonthlyCharges'].median()).astype(int)
```

```python
binary_cols = ['Partner','Dependents','PhoneService','PaperlessBilling']

for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})
```

```python
df = pd.get_dummies(df, drop_first=True)
```
